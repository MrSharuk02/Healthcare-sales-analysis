# 3. Exploratory Data Analysis(EDA)
## 3.1 Demographic Analysis:

```python
# Demographic Analysis
age_distribution = df['Age'].value_counts()
gender_distribution = df['Gender'].value_counts()
blood_type_distribution = df['Blood Type'].value_counts()

# Plotting Age Distribution
plt.figure(figsize=(10, 6))
plt.bar(age_distribution.index, age_distribution.values, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Age Distribution of Patients')
plt.grid(True)
plt.show()
```

```python
# Plotting Gender Distribution
plt.figure(figsize=(6, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Gender Distribution of Patients')
plt.show()
```

```python
# Plotting Blood Type Distribution
plt.figure(figsize=(8, 6))
blood_type_distribution.plot(kind='bar', color='lightgreen', ylim= (1200,1300))
plt.xlabel('Blood Type')
plt.ylabel('Number of Patients')
plt.title('Blood Type Distribution of Patients')
plt.grid(True)
plt.show()
```

```python
# Examine trends in admission types based on age
age_admission_trends = df.groupby(['Age', 'Admission Type'])['Name'].count().unstack()
age_admission_trends.plot(kind='bar', stacked=True, figsize=(12, 8),legend='k')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Trends in Admission Types based on Age')
plt.xticks(rotation=90)
plt.legend(title='Admission Type')
plt.grid(True)
plt.show()
```

```python

# Examine trends in admission types based on gender
gender_admission_trends = df.groupby(['Gender', 'Admission Type'])['Name'].count().unstack()
gender_admission_trends.plot(kind='bar', stacked=False, figsize=(8, 6), edgecolor='k')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.title('Trends in Admission Types based on Gender')
plt.xticks(rotation=0)
plt.legend(title='Admission Type')
plt.grid(True)
plt.show()
```

```python


# Create a box plot for age distribution across admission types
plt.figure(figsize=(10, 6))
df.boxplot(column='Age', by='Admission Type', figsize=(10, 6))
plt.xlabel('Admission Type')
plt.ylabel('Age')
plt.title('Age Distribution Across Admission Types')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

# Count the occurrences of each medical condition
medical_condition_counts = df['Medical Condition'].value_counts()

# Display the top 10 most common medical conditions
top_10_medical_conditions = medical_condition_counts.head(10)
print("Top 10 Most Common Medical Conditions:")
print(top_10_medical_conditions)
```

```python

# Frequency analysis of medical conditions by age group
medical_condition_by_age = df.groupby(['Age', 'Medical Condition']).size().unstack(fill_value=0)

# Plotting
plt.figure(figsize=(30,15))
medical_condition_by_age.plot(kind='bar', stacked=1)
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.title('Medical Conditions by Age Group')
plt.xticks(rotation=90)
plt.legend(title='Medical Condition')
plt.show()
```
